```python
import pandas as pd 
import numpy as np 
import matplotlib.pylab as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
%matplotlib inline
pd.set_option('display.max_columns', 200)
```

```python
## Reading the data 
january = pd.read_parquet('data/yellow_tripdata_2026-01.parquet')
february = pd.read_parquet('data/yellow_tripdata_2026-02.parquet')
```

```python
data = pd.concat([january, february], axis=0).reset_index()
```

```python
print(data.shape)
data.head()
```

```python
data.hist(bins=100, figsize=(25, 15))
plt.show()
```

### Data Cleaning

```python
data.describe(include='all')
```

```python
correlation_matrix = data.drop('store_and_fwd_flag', axis= 1).corr()
correlation_matrix
```

```python
usefuldata = data.copy()
```

```python
usefuldata = usefuldata[['index', 'tpep_pickup_datetime', 'tpep_dropoff_datetime', 'passenger_count', 'trip_distance', 
                         'RatecodeID', 'store_and_fwd_flag', 'PULocationID', 'DOLocationID', 'payment_type', 
                         'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 
                         'total_amount', 'congestion_surcharge', 'Airport_fee',]]

usefuldata
```

```python
round(usefuldata.isnull().sum() / len(usefuldata) * 100)
```

```python
usefuldata['passenger_count'].hist(bins=20, figsize=(10, 5))
plt.show()
```

```python
usefuldata[usefuldata.passenger_count > 1].reset_index()
```

```python
usefuldata = usefuldata.drop(usefuldata[usefuldata.passenger_count == 0].index)
```

```python
usefuldata = usefuldata.drop(usefuldata[usefuldata.passenger_count> 1].index).reset_index()
```

```python
usefuldata[usefuldata['passenger_count'].isna()].drop('level_0', axis=1).reset_index()
```

```python
usefuldata = usefuldata[usefuldata.passenger_count == 1].drop('index', axis=1).reset_index()
```

```python
usefuldata = usefuldata.set_index('index')
```

```python
usefuldata.columns.values
```
