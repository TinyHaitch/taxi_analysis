## Creating a BaseModel

##### 1. Using Decision Tree Regressor

```python
features = ['passenger_count', 'PULocationID', 'DOLocationID']
target = 'trip_distance'
```

```python
from sklearn.model_selection import train_test_split

X_bench = final_data[features]
y_bench = final_data[target]

X_bench = pd.get_dummies(X_bench)

X_train, X_test, y_train, y_test = train_test_split(X_bench, y_bench, train_size=0.35, random_state=21)
```

```python
from sklearn.tree import DecisionTreeRegressor

reg_model_base = DecisionTreeRegressor(max_depth=10,
                                   criterion='squared_error',
                                  splitter='random')
reg_model_base.fit(X_train, y_train)
```

```python
# print("\n" + "="*60 + "\n")
# print("                    UNIVARIATE ANALYSIS")
# print("\n" + "="*60 + "\n")
```

```python
y_pred_base = reg_model_base.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from math import sqrt

print(f'Mean Squared Error: {mean_squared_error(y_pred=y_pred_base, y_true=y_test)}')
print(f'Mean Absolute Error: {mean_absolute_error(y_pred=y_pred_base, y_true=y_test)}')
print(f'R2 Score: {r2_score(y_pred=y_pred_base, y_true=y_test)}')
print(f'Root Mean Squared Error: {sqrt(mean_squared_error(y_pred=y_pred_base, y_true=y_test))}')
```

```python
res1 = {'y_pred': y_pred_base, 'y_true': y_test}

res_frame1 = pd.DataFrame(res1)
```

```python
res_frame1.plot(kind='scatter', x='y_pred', y='y_true', figsize=(10, 5))
plt.show()
```

## Feature Engineering

```python
new_useful_data = usefuldata.copy()
```

```python
new_useful_data
```

```python
## Creating different time from pickup time

new_useful_data['transaction_date'] = pd.to_datetime(new_useful_data['tpep_pickup_datetime'].dt.date)
new_useful_data['transaction_year_pick'] = new_useful_data['transaction_date'].dt.year
new_useful_data['transaction_month_pick'] = new_useful_data['transaction_date'].dt.month
new_useful_data['transaction_day_pick'] = new_useful_data['transaction_date'].dt.day
new_useful_data['transaction_day_name_pick'] = new_useful_data['transaction_date'].dt.day_name()


## Creating different time from dropoff time 

new_useful_data['transaction_dropoff_date'] = pd.to_datetime(new_useful_data['tpep_dropoff_datetime'].dt.date)
new_useful_data['transaction_year_dropoff'] = new_useful_data['transaction_dropoff_date'].dt.year
new_useful_data['transaction_month_dropoff'] = new_useful_data['transaction_dropoff_date'].dt.month
new_useful_data['transaction_day_dropoff'] = new_useful_data['transaction_dropoff_date'].dt.day 
new_useful_data['transaction_day_name_dropoff'] = new_useful_data['transaction_dropoff_date'].dt.day_name()
```

```python
new_useful_data.columns.to_list
```

```python
## Converting columns to neccessacy datatype 
new_useful_data['PULocationID'] = new_useful_data['PULocationID'].astype('str') 
new_useful_data['DOLocationID'] = new_useful_data['DOLocationID'].astype('str')
new_useful_data['RatecodeID'] = new_useful_data['RatecodeID'].astype('str')
```

```python
## Data for model 2 
input_feature = ['RatecodeID','PULocationID', 'DOLocationID', 'transaction_month_pick', 'transaction_day_pick', 
                 'transaction_day_name_pick','transaction_year_dropoff', 'transaction_month_dropoff', 
                 'transaction_day_dropoff', 'transaction_day_name_dropoff']

label = 'total_amount'
```

```python
from sklearn.model_selection import train_test_split

X_features1 = new_useful_data[input_feature]
y_label1 = new_useful_data[label]

## Binnig my X 
X_features1 = pd.get_dummies(X_features1)

X_train1, X_test1, y_train1, y_test1 = train_test_split(X_features1, y_label1, train_size=0.35, random_state=21)
```

```python
from sklearn.tree import DecisionTreeRegressor

reg_model1 = DecisionTreeRegressor(max_depth=10,
                                  criterion='squared_error', 
                                  splitter='random')

reg_model1.fit(X_train1, y_train1)
```

```python
y_pred1 = reg_model1.predict(X_test1)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from math import sqrt

print(f'Mean Squared Error: {mean_squared_error(y_pred=y_pred1, y_true=y_test1)}')
print(f'Mean Absolute Error: {mean_absolute_error(y_pred=y_pred1, y_true=y_test1)}')
print(f'R2 Score: {r2_score(y_pred=y_pred1, y_true=y_test1)}')
print(f'Root Mean Squared Error: {sqrt(mean_squared_error(y_pred=y_pred1, y_true=y_test1))}')
```

```python
data_res1 = {'y_true':y_test1, 'y_pred':y_pred1}

data_res1 = pd.DataFrame(data_res1)
data_res1
```

```python
data_res1.plot(kind='scatter', x='y_true', y='y_pred', figsize=(10, 5))
plt.show()
```
